## 1. Introduction

Plug flow tubular reactors (PFTRs) are widely used in the chemical industry for continuous production processes because they provide efficient conversion and allow good control over reaction conditions. Compared to batch reactors, PFTRs operate under steady-state conditions and are particularly suitable for large-scale applications where heat transfer and residence time strongly influence reactor performance. For exothermic reactions, temperature control becomes especially important, since heat released during the reaction can significantly affect reaction rates and process stability.

In this work, the hydrolysis of ethylene oxide (EO) with water to produce ethylene glycol (EG) is investigated as an example of an exothermic reaction system. Ethylene glycol is an industrially important chemical used in applications such as antifreeze products and polyester manufacturing. Since the reaction releases heat, reactor operation depends not only on reaction kinetics but also on thermal effects and operating conditions.

The aim of this study is to analyze the behavior of the reaction system in a plug flow tubular reactor by investigating the influence of relevant process parameters. Parameters such as heat transfer, inlet temperature, reactant concentration, and flow-related conditions are evaluated to better understand their impact on reactor temperature and reaction behavior. In addition, the suitability of the reactor concept for the considered reaction system is discussed.


## 2. Mathematical Model

For the present work, the hydrolysis of ethylene oxide (EO) to ethylene glycol (EG) was modeled using a plug flow tubular reactor (PFTR). Since a PFTR can be represented as an initial value problem through the residence time concept, the reactor model was solved using a time-based formulation.

The residence time was calculated according to:


$$
\tau = \frac{L}{u}
$$


where (L) is the reactor length and (u) is the fluid velocity. Using the given reactor length of 300 m and a fluid velocity of 1 m s(^{-1}), a residence time of 300 s was obtained.

The considered reaction is:


$$
EO + H_2O \rightarrow EG
$$


The reaction kinetics were described using an Arrhenius-type expression:


$$
k(T)=5.01 \cdot \exp \left(-\frac{72500}{RT}\right)
$$


Assuming first-order dependence on ethylene oxide concentration, the reaction rate was calculated as:


$$
r = k(T)c_{EO}
$$


To describe the reactor behavior, component balances and an energy balance were implemented to account for concentration changes and temperature evolution due to reaction heat and heat transfer through the reactor wall.

## 4. Model Implementation

To investigate the reactor behavior under different operating conditions, the mathematical model was numerically implemented and solved. The simulation framework was developed using Python in a Jupyter Notebook environment, allowing efficient implementation of the governing equations and visualization of the obtained results.

The reactor model was formulated as a coupled system of ordinary differential equations (ODEs) describing concentration and temperature changes throughout the reaction process. Since the plug flow tubular reactor (PFTR) was represented using the residence time concept, the governing equations were solved as an initial value problem over the calculated reactor residence time.

For numerical calculations, the NumPy and SciPy libraries were used, while graphical representation of the obtained reactor behavior was performed using Matplotlib. The `solve_ivp` function together with the BDF numerical solver was applied to solve the system of differential equations, since it provides stable numerical performance for reaction systems with coupled mass and energy balances.

Furthermore, sensitivity analyses were carried out to investigate the effect of important process variables such as inlet temperature, reactant concentration, and heat-transfer conditions on reactor performance. The obtained concentration and temperature profiles were subsequently analyzed to evaluate the thermal and kinetic behavior of the system.
