# Assignment 4: Backmixing in tubular reactors

## Introduction and Objective

Plug Flow Reactor (PFR) is a theoretical model that is ideal in which there is no axial mixing. This model present a simplified design and performance analysis of a reactor, but it does not account for non-ideal flow behaviour that is often found in real reactors. The mixing of fluid elements in the axial direction in real tubular reactors is known as axial dispersion or backmixing. This non-ideal mixing has the effect to change the residence time distribution of the reacting species, which is important to determine the conversion, selectivity and overall efficiency of the reactor. Therefore, a reliable modelling of axial dispersion is crucial to the design, optimization and scale up of industrial reactors [1][2].

The axial dispersion is especially significant when continuous tubular reactors are used in industrial applications. An example of such processes is the production of ethylene oxide, in which ethylene is selectively oxidized on silver catalysts in multitubular fixed bed reactors. For these reactors, the effects of axial dispersion on the concentration and temperature profiles are relevant, and these directly impact the selectivity, yield and safety of the reactor. To predict the behaviour of the reactor and optimize the operating conditions for large-scale industrial production, reliable models of the reactors are needed [3].

The axial dispersion model is one of the most popular of the models used to describe non-ideal flow behaviour. In this model, both convective transport and axial mixing are taken into account and more realistic performance of a tubular reactor can be obtained than for the ideal PFR model. The Bodenstein number (Bo), defined as the ratio of convective transport to axial dispersion is an important dimensionless number in the model. A high Bodenstein number means that backmixing is small and the reactor is closer to ideal plug flow behaviour, while a low Bodenstein number means that the axial mixing is larger, and the reactor has more CSTR behaviour. Typically, the governing equations are solved using Boundary Value Problem (BVP) formulations with the Danckwerts boundary conditions, which are physically consistent description of the mass transport at the boundary of the reactor inlet and outlet [2].

The behaviour of a tubular reactor of length 10 m and inner diameter 32 mm is investigated, operating at 1 L/s of volumetric flow rate in the reactor and at a reactant A inlet concentration of 100 mol/m³. Both Boundary Value Problem (BVP) and Initial Value Problem (IVP) formulations are used for solving the governing equations of the axial dispersion model. In addition, the effect of the Bodenstein number on the concentration profile is studied and the results of the axial dispersion model are compared to those of an equivalent cascade of Continuous Stirred Tank Reactors (CSTRs) to determine the extent to which the former can be used to simulate non-ideal reactor behaviour.
