Thus the following task should be solved:\
In an isothermal BSTR the following reaction takes place at 235°C:
$$A \xrightarrow{k_1} B$$
The reaction is described by the following system of ODEs:
$$\frac{dC_A}{dt} = -k C_A^{n_1}$$
The initial conditions are: $C_A(0) = 1 \, \mathrm{mol/L}$, $C_B(0) = 0 \, \mathrm{mol/L}$.
The kinetic constant $k$ is unknown.

Question: Find the kinetic konstant with the provided experimental data.


Solution approach:
To solve this problem, we will follow these steps:
1. Define a function for the reaction rate and the system of ODEs.
2. Use `scipy.integrate.solve_ivp` to solve the system of ODEs for a chosen kinetic constant.
3. Plot the concentration profiles of A and B.
4. Import the provided experimental data.
4. Set up an optimization problem using `lmfit.Minimizer` to find the reaction rate constant.
